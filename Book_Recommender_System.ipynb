Popularity Based Recommender System

```python
books_with_ratings = ratings.merge(books, on='ISBN')
```

```python
books_with_ratings.shape
```

```python
books_with_ratings
```

```python
books_with_ratings.drop(["Image-URL-S"], axis=1, inplace=True)
```

```python
books_with_ratings.drop(["Image-URL-M"], axis=1, inplace=True)
```

```python
books_with_ratings.drop(["Image-URL-L"], axis=1, inplace=True)
```

```python
x = books_with_ratings.groupby('User-ID').count()['Book-Rating'] > 200
x[x]
```

```python
max_users = x[x].index
```

```python
filtered_ratings = books_with_ratings[books_with_ratings['User-ID'].isin(max_users)]
```

```python
y = filtered_ratings.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
famous_books
```

```python
final_ratings = filtered_ratings[filtered_ratings['Book-Title'].isin(famous_books)]
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
def recommend(book_name):
  index=np.where(pt.index==book_name)[0][0]
  similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5]

  for i in similar_items:
      print(pt.index[i[0]])
```

```python
recommend('The Vampire Lestat (Vampire Chronicles, Book II)')
```
